```python
class perceptron:
    def __init__(self, name :str):
        self.name = name
        length = 36608
        self.w= list()
        for i in range(length):
            self.w.append(np.random.rand())
        
        self.b = np.random.rand()
    
def calc_z(x,w,b):
    c = 0
    for i in range(len(x)):
        c += x[i] * w[i]
    return c + b

def threshold( c):
    if c > 0:
        return 1
    else:
        return 0

def cost(x,y, w, b):
    a = calc_z(x,w,b)
    #print(a)
    a = threshold(a)
    return y-a

def train(error, w: list, b):
    learning_rate = .01
    a = list()
    for i in range(len(w)):
        a.append(w[i] + (learning_rate * error))
    b1 = b + (learning_rate* error)
    
    return a,b1
    
    
        
```

```python
mild_perceptron = perceptron('mild')
moderate_perceptron = perceptron('moderate')
non_perceptron = perceptron('non')
verymild_perceptron = perceptron('very mild')
```

```python
w = list()
length = 36608
for i in range(length):
    w.append(np.random.rand())
w = np.array(w)

b = np.random.rand()
```

```python
cateogies = ['MildDemented',
 'ModerateDemented',
 'NonDemented',
 'VeryMildDemented']
```

```python
for c in cateogies:
    print(len(os.listdir(train_path+ '/' + c)))
```

```python
y = [1,0,0,0]
```

```python
def predict(X:list, weights:list, b: int):
    y_pred = (np.dot(weights.T, X) + b)
    y_pred = threshold(y_pred)
    return y_pred
```

<h1> Training the data </h1><br>
Epochs = 10<br>
Training Size per Category = 40

```python
learning_rate = .25
p = 0

miss_classify = 0
total_num = 0

for iteration in range(10):
    it=0
    e_rate = 0
    for c in cateogies:
        error = 0
        print('===='*10)
        print(c)
        size = (len(os.listdir(train_path+ '/' + c)))
        for i in range(40):
            error = 0
            temp = os.listdir(train_path + '/' + c)[i]
            temp = train_path + '/' + c +'/' + temp
            temp = imread(temp)
            temp = temp.reshape(36608)
            temp = temp/255

            a = predict(temp,w,b)
            error1 = (y[it]-a)

            if (error1 !=0):
                miss_classify += 1
                e_rate += 1
            print(f'a: {a} y: {y[it]} error: {error1}')
            error += error1
            total_num += 1
        
        for j in range(36608):
            w[j] = w[j] + (learning_rate * error * temp[j])
            #print(temp[j])
            
        b += (learning_rate * error)
        it += 1
        
print(miss_classify)
```

```python
training_accuracy = 1 - (360/1600)
training_accuracy
```
